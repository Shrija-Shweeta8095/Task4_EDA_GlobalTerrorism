```python
## rename the columns
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':
                  'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                  'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group',
                   'targtype1_txt':'TargetType','weaptype1_txt':'Weapon_type','motive':'Motive',
                  'extended':'Extended'},inplace=True)
```

```python
df1=df[['Year','Month','Extended','Day','Country','State','Region','city','latitude','longitude',
       'AttackType','Killed','Wounded','Target','Summary','Group','TargetType','Weapon_type','Motive']]
```

```python
df1.head()
```

```python
## check null values
df1.isnull().sum()
```

```python
df1.info()
```

```python
## DESTRUCTIVE FEATURES
print('Country with the most attacks:',df['Country'].value_counts().idxmax())
print('City with the most attacks:',df['city'].value_counts().index[1])
print('Region with the most attacks:',df['Region'].value_counts().idxmax())
print('Year with the most attacks:',df['Year'].value_counts().idxmax())
print('Month with the most attacks:',df['Month'].value_counts().idxmax())
print('Group with the most attacks:',df['Group'].value_counts().index[1])
print('Most attack types:',df['AttackType'].value_counts().idxmax())
```

```python
## Terrorist Activities by Region in each Year through Area Plot
pd.crosstab(df1.Year,df1.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
## Number of Terrorist Activities each year
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df1,palette='RdYlGn_r',edgecolor=sns.color_palette('YlOrBr',10))
plt.xticks(rotation=90)
plt.title('Number of Terrorist Activities Each Year')
plt.show()
```

Most of the Terrorist actvities happened in the year 2014

```python
## History of the worst Terror Attacks in Heatmap using Plotly
df1['Wounded']=df1['Wounded'].fillna(0).astype(int)
df1['Killed']=df1['Killed'].fillna(0).astype(int)
df1['casualities']=df1['Killed']+df1['Wounded']
```

```python
## values are sorted by the top 40 worst terror attacks as to keep the heatmap simple and eady to visualize
df2=df1.sort_values(by='casualities',ascending=False)[:40]
```

```python
heat=df2.pivot_table(index='Country',columns='Year',values='casualities')
heat.fillna(0,inplace=True)
```

```python
plt.figure(figsize=(30,5))
sns.countplot('Country',data=df1)
plt.xticks(rotation=90)
plt.show()
```

Maximum terrorist activities has been done in Iraq.

```python
plt.figure(figsize = (10,5))
df1.groupby(['Country'])['Killed'].sum().sort_values(ascending = False).head(15).plot(kind = 'bar', color = 'pink')

plt.xticks(rotation=90)
plt.ylabel("Killed")
plt.show()
```

Most of the people of Iraq has been killed in Terrorism.
